## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
## Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")
```

```python
#Each mouse should* be measured at time (0,45,5) 0 to 45 days in incriments of 5 (this is not the case)
merge_df.head()
```

```python
## Checking the number of mice in the DataFrame.

#Made an array to store the names of each mouse, then measured the length
#The instructions note that this was a study of 249 mice, and that is what I found
UniqueMice = merge_df['Mouse ID'].unique()
print(f'The number of individual mice studied was: {len(UniqueMice)}')
```

```python
## Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

# Note: this gives me a df for all duplicate rows after the first occurance
DupMouseCheck = merge_df[merge_df.duplicated(['Mouse ID', 'Timepoint'])]
```

```python
## Optional: Get all the data for the duplicate mouse ID. 
DupMouseCheck
```

```python
## Create a clean DataFrame by dropping the duplicate mouse by its ID.

#Note I chose to drop the duplicate row after the first occurance. 
CleanMouse = merge_df.drop_duplicates(subset=['Mouse ID','Timepoint'], keep="first", inplace=False)
```

## Summary Statistics

```python
## Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
## This method is the most straighforward, creating multiple series and putting them all together at the end.
```

```python
#group data by Drug Regimen & Timepoint (The addition of timepoint in grouped data will be for bar graph later)
DrugReg = CleanMouse.groupby(['Drug Regimen','Timepoint'])
```

```python
#Created an empty df, I added count as a column
DrugRegSum = pd.DataFrame(columns = ['Count','Mean','Median','Variance','SD','SEM'])
```

```python
#Populate the empty df with summary statistics
DrugRegSum['Count'] = DrugReg['Tumor Volume (mm3)'].count()
DrugRegSum['Mean']=DrugReg['Tumor Volume (mm3)'].mean()
DrugRegSum['Median']=DrugReg['Tumor Volume (mm3)'].median()
DrugRegSum['Variance']=DrugReg['Tumor Volume (mm3)'].var()
DrugRegSum['SD']=DrugReg['Tumor Volume (mm3)'].std()
DrugRegSum['SEM']=DrugReg['Tumor Volume (mm3)'].sem()
```

## Quartiles, Outliers and Boxplots

```python
## Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. 
```

```python
#Gets only the 45th day measurements
#This gives the last measurment for each mouse (final tumor volume)
#This will drop all mice that died before the last measurment
ResultMouse = CleanMouse[CleanMouse['Timepoint'] == 45]
```

```python
ResultMouse.head()
```

```python
#Find out which were the top performing drugs. 
#Criteria = most mice who survived to 45 days & Lowest mean tumor size

#Create empty df to hold results
ResultMouseSum = pd.DataFrame(columns =['Count 45th Day','Mean Tumor Size'])

#Create group the df by drug type
ResultMouse_bydrug = ResultMouse.groupby('Drug Regimen')

#Get the count of mice at 45 days per drug
ResultMouseSum['Count 45th Day'] = ResultMouse_bydrug['Mouse ID'].count()

#Get the mean tumor size for mice that lived to 45 days
#Assumes we don't care about a small tumor size in a mouse that died prematurely 
ResultMouseSum['Mean Tumor Size'] = ResultMouse_bydrug['Tumor Volume (mm3)'].mean()
```

```python
#Sort the results highest survival to lowest
Survival = ResultMouseSum.sort_values(by='Count 45th Day', ascending=False)

#Sort the results lowest tumor size to highest
TumorSize = ResultMouseSum.sort_values(by='Mean Tumor Size', ascending=True)

#Take the top 4 performing drugs on each measure (the first 4 rows after the sort)
Survival_Top4 = Survival[0:4]
TumorSize_Top4 = TumorSize[0:4]
```

```python
#These lists ARE NOT identical. One potential explination is that the effectiveness of drugs
# like Infubinol are not evenly distributed. Some mice may respond very well while others do not, 
# resulting in higher mortality but also better results in those that survive. 
# Perhaps some drugs have a higher rate of leathal side effects. 
# This brings up questions about how to define 'top performing'

# I have chosen to use those drugs with the highest number of mice surviving after 45 days
print(Survival_Top4)
print(TumorSize_Top4)
```

```python
#Get a list of the top4 drug names
Top4list_Survival = list(Survival_Top4.index.get_level_values(0))

#Subset the original df, only get mice who were given the top performing drugs
ResultMouse_Top4 = ResultMouse[ResultMouse['Drug Regimen'].isin(Top4list_Survival)]
```

```python
#Sort that df according to the drug, and then the tumor size
ResultMouse_Top4 = ResultMouse_Top4.sort_values(['Drug Regimen','Tumor Volume (mm3)'])
ResultMouse_Top4
```

```python
##Calculate the IQR and quantitatively determine if there are any potential outliers.
## data frame TV_IQR stands for tumor volume IQR. This was only for top performing drugs
TV_IQR = ResultMouse_Top4['Tumor Volume (mm3)'].quantile([.25,.5,.75])
lowerq = TV_IQR[0.25]
upperq = TV_IQR[0.75]
iqr = upperq-lowerq
```

```python
print(f"The lower quartile of Tumor Volume is: {lowerq}")
print(f"The upper quartile of Tumor Volume is: {upperq}")
print(f"The interquartile range of Tumor Volume is: {iqr}")
print(f"The the median of Tumor Volume is: {TV_IQR[0.5]} ")
```

```python
#Note, the interquartile range is very large. 
#This creates impossible scenarios for outliers
#Like tumor volume being below 0. 

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
#Look to see if there are any outliers 
outlier_Tumor = ResultMouse_Top4.loc[(ResultMouse_Top4['Tumor Volume (mm3)'] < lower_bound) | (ResultMouse_Top4['Tumor Volume (mm3)'] > upper_bound)]
```

```python
#No outliers showing up
outlier_Tumor
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
MouseWeight = Mouse_Capomulin['Weight (g)']
MouseTumor = Mouse_Capomulin['Tumor Volume (mm3)']
```

```python
MouseCorr = round(st.pearsonr(MouseWeight,MouseTumor)[0],2)
```

```python
print(f"The correlation coefficient between mouse weight and tumor volume is {MouseCorr}")
```

```python
mwt_slope, mwt_int, mwt_r, mwt_p, mwt_std_err = st.linregress(MouseWeight, MouseTumor)
```

```python

print(f'The weight of the mouse was significantly related to the tumor volume {mwt_p}')
print(f'Weight can explain {round((mwt_r*mwt_r)*100,2)}% of the variation in tumor volume')
```

```python
#Creating a regression line
mwt_fit = mwt_slope * MouseWeight + mwt_int
```

```python
plt.scatter(Mouse_Capomulin['Weight (g)'],Mouse_Capomulin['Tumor Volume (mm3)'], c=Mouse_Capomulin['Timepoint'])
clb = plt.colorbar()
#Adding the regression line to the plot
plt.plot(MouseWeight,mwt_fit,"--")
plt.xlabel("Weight")
plt.ylabel("Tumor Volume")
clb.set_label("Timepoint")

plt.suptitle('Regression of Tumor Volume by Weight for All Mice on Capomulin', fontsize=12)
plt.savefig("/Users/agetecza/my-data-class/matplotlib-challenge/Pymaceuticals/Figures/MouseRegress.png")
```
